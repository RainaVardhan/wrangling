**Q1 Answers**
1. Reading the abstract, this paper is probably going to talk about data tidying and how its framework makes it easier to facilitate manipulation, modeling, and visualization on datasets.
2. The tidy data standard is intented to make data cleaning easier in the initial stages as it provides the development of tools that allow you to directly focus on the domain problem and analyzing the data without having to start from square one.
3. The saying, "Like families, tidy datasets are all alike but every messy dataset is messy in its own way" means that all tidy share the same structure making them alike while messy datasets all have their own problems that each require a unique specific approach to clean and organize. The other saying "For a given dataset, it's usually easy to figure out what are observations and what are variables, but it is surprisingly difficult to precisely define variables and observations in general" means that though identifying what the rows and columns in a datatset are on their own is usually straightforward, figuring our what they mean in relation to the entire dataset and their universal application is a lot harder to disect and establish.
4. Values are either numbers or strings that belong to a variable and observation. Variables measure the same attribute across all units. Observations measure the same unit across attributes. Both the variables and observations contain all the values.
5. Tidy data is defined as being "a standard way of mapping the meaning of a dataset to its structure," meaning that restructures and arranges the data in a way that conveys what the data inherently means clearly. In tidy data, "each variable forms a column, each observation forms a row, and each type of observation unit forms a table".
6. The 5 most common problens with messy datasets are that their "colunn headers are values, not variable names, multiple variables are stored in one column, variables are stored in both rows and columns, multiple types of observational units are stored in the same table, and a single observational unit is stored in multiple tables". The data in table 4 is messy because column headers are values not variable names as income ranges are numbers meaning they shouldn't be column headers. Melting a dataset means that you turn the columns into rows because the columns variables are values so rearranging them this way allows for a more clear understanding of the data.
7. Table 11 is messy because some of the variable names are actually values as the days of the month (d1-d31) are column names when they should be values. Table 12a is molten data because they at least molted all the datas under the column name date but the element column contains variable names tmax and tmin instead of values so its not quite tidy. Table 12b is tidy because they molted the dates under the column name date and tmax and tmin are variables now so all the variables contain values.
8. The chicken-and-egg problem with tidying data is that "if tidy data is only as useful as the tools that work with it, then tidy tools will be inextricably linked to tidy data" meaning that the efficiency and value of working with tidy data only comes with the availability of tools specifically design to handle and analyze the datasets in this tidy form. Wickham hopes that in the future, the tidy framwork will be further developed to allow for better data storage strategies and better tools to make tasks involved with cleaning data easier.

```python
#Q1.1
import numpy as np
import pandas as pd
df = pd.read_csv('/content/wrangling/assignment/data/airbnb_hw.csv', low_memory = False)
print(df.shape,'\n')
print(df.dtypes,'\n')
df.head()
```

```python
price = df['Price'] #Creating a copy of the Price colomn
price.unique()
```

```python
price = pd.to_numeric(price.str.replace(',','')) #replaceing all the commas with empty character and then forcing it to the numeric datatype
price.unique()
```

```python
df['Price'] = price #Assigning the Price column with the now cleaned data
print('Total Missings: \n', sum(df['Price'].isnull()),'\n')
```

```python
#Q2.2
df = pd.read_csv('/content/wrangling/assignment/data/sharks.csv', low_memory = False)
print(df.shape,'\n')
print(df.dtypes,'\n')
df.head()
```

```python
type = df['Type'] #creating a copy of the Type column
type.unique()
```

```python
type = type.replace(['Boating', 'Sea Disaster', 'Watercraft', 'Boatomg'],'Boat') #Combining all the boat like types under the column name Boat
type = type.replace([np.nan,'Unverified','Questionable','Under investigation','Invalid'],'Unconfirmed') #Combining all the unconfirmed or empty types under the column name Unconfirmed
type.unique()
```

```python
type.value_counts()
```

```python
#Q2.3
df = pd.read_parquet('/content/wrangling/data/justice_data.parquet')
print(df.shape,'\n')
print(df.dtypes,'\n')
df.head()
```

```python
wasReleased = df['WhetherDefendantWasReleasedPretrial'] #making a copy of the column 'WhetherDefendantWasReleasedPretrial'
wasReleased.unique()
```

```python
wasReleased = wasReleased.replace(9,np.nan) #Replacing the 9s with nans because in the codebook it says that 9 means unclear
wasReleased.unique()
```

```python
df['WhetherDefendantWasReleasedPretrial'] = wasReleased #assigning the 'WhetherDefendantWasReleasedPretrial' column to the newly cleaned data
df['WhetherDefendantWasReleasedPretrial'].value_counts()
```

```python
#Q2.4
sentenceType = df['SentenceTypeAllChargesAtConvictionInContactEvent'] #making a copy of the column 'SentenceTypeAllChargesAtConvictionInContactEvent'
imposedSentence = df['ImposedSentenceAllChargeInContactEvent'] #making a copy of the column 'ImposedSentenceAllChargeInContactEvent'
imposedSentence.unique()
```

```python
imposedSentence = pd.to_numeric(imposedSentence, errors='coerce') #converting all the values to numeric so that we can analyze it better
imposedSentence.unique()
```

```python
isNull = imposedSentence.isnull() #creating a separate variable that will hold all the missing values
print(np.sum(isNull),'\n')
pd.crosstab(isNull,sentenceType) #Checking how the missing values relate to the sentenceType values to see patterns
```

```python
imposedSentence = imposedSentence.mask(sentenceType == 4, 0) #changed the imposedSentence value to 0 for all sentenceType values that are equal to 4 because type 4 in the codebook means the charges were pending, dismissed, or deferred meanign that there was no sentence
imposedSentence = imposedSentence.mask(sentenceType == 9, np.nan) #changed the imposedSentence value to nan for all sentenceType values that are equal to 9 because type 9 in the codebook means non applicable
isNull = imposedSentence.isnull() #creating a separate variable that will hold all the missing values after we cleaned some of the data
pd.crosstab(isNull,sentenceType) #checking to see if all the missing values for sentenceType 4 were updated.
```

```python
print(np.sum(isNull),'\n')
```

```python
df['SentenceTypeAllChargesAtConvictionInContactEvent'] = sentenceType #making 'SentenceTypeAllChargesAtConvictionInContactEvent' equal to the newly cleaned data in sentenceType
df['ImposedSentenceAllChargeInContactEvent'] = imposedSentence #making 'ImposedSentenceAllChargeInContactEvent' equal to the newly cleaned data imposedSentence
```
